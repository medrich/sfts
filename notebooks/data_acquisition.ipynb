```python
# imports

from bs4 import BeautifulSoup
import requests
import sys
from collections import defaultdict
import pickle
import os
import json
```

```python
# get EveryNoise HTML

every_noise = requests.get('https://everynoise.com/engenremap.html')
```

```python
soup = BeautifulSoup(every_noise.text, 'html.parser')
genre_tags = soup.find_all('div', 'genre scanme')
```

```python
# traverse genre_tags and navigate to each page

genres = defaultdict(list)
genre_count = 1

for tag in genre_tags:
    print(f'Gathering musical insight #{genre_count}')
    
    genre = tag.getText()
    print(genre)
    genre_clean = genre[:-2]
    print(genre_clean)
    page = 'https://everynoise.com/engenremap-' + str.replace(genre_clean, " ", "") + '.html'
    print(page)
    
    # while on that page, get the artists
    genre_artists = requests.get(page)
    soup2 = BeautifulSoup(genre_artists.text, 'html.parser')
    artist_tags = soup2.find_all('div', 'genre scanme')
    
    for tag in artist_tags:
        artist = tag.getText()
        artist_clean = artist[:-2]
        print(artist_clean)
        genres[genre_clean].append(artist_clean)
            
    genre_count += 1
```

```python
with open('genres.p', 'wb') as fp:
    pickle.dump(genres, fp)
```

```python
with open('genres.p', 'rb') as fp:
    genres = pickle.load(fp)
```

```python
print(f'Found {len(genres)} different genres.')
```

```python
# connect each artist to all the genres they qualify for

artists = defaultdict(list)

for genre in genres:
    bands = genres[genre]
    
    for b in bands:
        artists[b].append(genre)

# get rid of empty keys
#del artists[""]

print(f'Found {len(artists)} artists in the data.')
```

```python
type(artists)
```

```python
# alphabetize lists and remove duplicates

for artist in artists:
    artists[artist] = list(set(artists[artist]))
    #artists[artist] = filter(None, artists[artist])
    artists[artist].sort()
for genre in genres:
    genres[genre] = list(set(genres[genre]))
    #genres[genre] = filter(None, genres[genre])
    genres[genre].sort()    
```

```python
# checking alphabeticity
#genres
```

```python
artists_clean, genres_clean = [], []

for artist in artists:
    artist = str(artist)
    if artist[0] == " ":
        artist = artist.replace(" ","", 1)
    artist = artist.replace('"', '')
    artists_clean.append(artist)
artists_clean.sort()

for genre in genres:
    genres_clean.append(genre)
genres_clean.sort()
```
